```python
#load all libraries

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import train_test_split
from sklearn.neighbors import KNeighborsClassifier
from sklearn.metrics import recall_score, precision_score
from sklearn.model_selection import cross_validate
from sklearn.model_selection import GridSearchCV
```

```python
cancer = pd.read_csv('dataset/wdbc.csv')
cancer
```

```python
cancer["diagnosis"]= cancer["diagnosis"].replace({
       "M" : "Malignant",
       "B" : "Benign"})
cancer
```

```python
standardized_cancer = cancer.copy()   #make a copy of the data
columns_to_exclude = ['id','diagnosis'] #columns we don't want to scale
columns_to_scale = standardized_cancer.columns.difference(columns_to_exclude)
```

```python
scaler = StandardScaler()   #initialize scaler
standardized_cancer[columns_to_scale] = scaler.fit_transform(standardized_cancer[columns_to_scale])
```

```python
standardized_cancer
```

```python
#set the seed
np.random.seed(1)

#split the data
cancer_train, cancer_test = train_test_split(standardized_cancer,train_size=0.75, shuffle=True, stratify=standardized_cancer["diagnosis"])
```

```python
cancer_train.info()
```

```python
cancer_test.info()
```

```python
# k = 5
#Step 1. Initialize our model
knn = KNeighborsClassifier(n_neighbors=5)
```

```python
#step 2. define our X and y variables (Predictor(X) and response(y))
X = cancer_train[["perimeter_mean","concavity_mean"]]
y = cancer_train["diagnosis"]
```

```python
#Step 3. fit model to train data
knn.fit(X,y)
```

```python
cancer_test["predicted"] = knn.predict(cancer_test[["perimeter_mean","concavity_mean"]])
cancer_test
```

```python
knn.score(cancer_test[["perimeter_mean","concavity_mean"]], cancer_test["diagnosis"])
```

```python
#confusion matrix
pd.crosstab(
    cancer_test["diagnosis"],
    cancer_test["predicted"]
)
```

```python
precision_score(
    y_true=cancer_test["diagnosis"],
    y_pred= cancer_test["predicted"],
    pos_label="Malignant"
)
```

```python
recall_score(
    y_true=cancer_test["diagnosis"],
    y_pred= cancer_test["predicted"],
    pos_label="Malignant"
)
```

```python
#creating a new test data
cancer_subtrain, cancer_validation = train_test_split(cancer_train,train_size=0.75,shuffle=True,stratify=cancer_train["diagnosis"])
```

```python
#initialize our model
knn = KNeighborsClassifier(n_neighbors=3)
X = cancer_subtrain[["perimeter_mean", "concavity_mean"]]
y = cancer_subtrain["diagnosis"]
knn.fit(X,y)
```

```python
knn.score(cancer_validation[["perimeter_mean","concavity_mean"]],cancer_validation["diagnosis"])
```

```python
#perform 5-fold cross validation
knn = KNeighborsClassifier(n_neighbors=3)
X = cancer_train[["perimeter_mean","concavity_mean"]]
y = cancer_train["diagnosis"]

returned_dictionary = cross_validate(
    estimator=knn,
    cv=5,    # setting up the cross validation number
    X=X,
    y=y
)
```

```python
returned_dictionary
```

```python
cv_5_df = pd.DataFrame(returned_dictionary)
cv_5_df
```

```python
cv_5_df.agg(["mean","sem"])
```

```python
#step 1: define our perimeter grid i.e. the values of K and by how it shall jump and the maximum value which is 100 -1 = 99
parameter_grid = {
    "n_neighbors" : range(1,100,5)
}
```

```python
#step 2. initialize our grid search CV object

cancer_tune_grid = GridSearchCV(
    estimator=knn,
    param_grid=parameter_grid,
    cv=10
)
```

```python
#step 3. Fit our grid search object to our training data

cancer_tune_grid.fit(cancer_train[["perimeter_mean","concavity_mean"]],cancer_train["diagnosis"])
```

```python
accuracy_grid = pd.DataFrame(cancer_tune_grid.cv_results_)
accuracy_grid
```

```python
# Create the plot
plt.figure(figsize=(10, 6))

# Plot mean test scores with error bars
plt.plot(accuracy_grid['param_n_neighbors'], accuracy_grid['mean_test_score'], '-o', color='blue')

# Add labels and legend
plt.xlabel('Number of Neighbors')
plt.ylabel('Accuracy estimate')
plt.title('K-Nearest Neighbors Performance')
plt.tight_layout()
plt.show()
```
